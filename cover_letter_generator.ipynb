```python
import google.generativeai as genai

API_KEY = "AQ.Ab8RN6LXesyEo1hngWgyI0eQNncQlPSLEWi-VUHdqMjoP527iQ"
genai.configure(api_key=API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
def generate_cover_letter(name, job_role, company, skills, experience):

    prompt = f"""
    Write a professional cover letter.

    Candidate Name: {name}
    Job Role: {job_role}
    Company: {company}
    Skills: {skills}
    Experience: {experience}

    The cover letter should:
    - Be professional and formal
    - Highlight relevant skills
    - Explain interest in the position
    - Be around 300-400 words
    """

    response = model.generate_content(prompt)
    return response.text
```

```python
name = input("Enter Your Name: ")
job_role = input("Enter Job Role: ")
company = input("Enter Company Name: ")
skills = input("Enter Skills (comma separated): ")
experience = input("Enter Experience: ")

cover_letter = generate_cover_letter(
    name,
    job_role,
    company,
    skills,
    experience
)

print("\nGenerated Cover Letter:\n")
print(cover_letter)
```

```python
with open("cover_letter.txt", "w", encoding="utf-8") as file:
    file.write(cover_letter)

print("Cover Letter saved as cover_letter.txt")
```
